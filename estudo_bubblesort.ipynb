```python
from re import A
import matplotlib.pyplot as plt
import time
import random
import matplotlib.pyplot as plt

def  bubble_sort(A):
  n = len(A)
  for i in range (n - 1):
    for j in range (n - 1):
      if A[j] > A[j + 1]:
        # troca os elementos
        temp = A[j]
        A[j] = A[j + 1]
        A[j + 1] = temp
  return A

# funcção para medir o tempo
def medir_tempo(A):
  tempo_inicio = time.time()
  bubble_sort(A)
  tempo_fim = time.time()
  return tempo_fim - tempo_inicio

# Dados de exemplo
tamanhos = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000]
tempos_execucao_desordenados = []
tempos_execucao_ordenados = []
# Medir o tempo de execução para diferentes tamanhos de entrada

for tamanho in tamanhos:
  # Teste com números desordenados
  A_desordenado = [random.randint(0, 1000) for _ in range(tamanho)]
  tempo_desordenado = medir_tempo(A_desordenado.copy())
  tempos_execucao_desordenados.append(tempo_desordenado)
  # Teste com números ordenados
  A_ordenado = list(range(tamanho))
  tempo_ordenado = medir_tempo(A_ordenado.copy())
  tempos_execucao_ordenados.append(tempo_ordenado)

# Plotar os gráficos
plt.plot(tamanhos, tempos_execucao_desordenados, marker='o', label='Desordenado')
plt.plot(tamanhos, tempos_execucao_ordenados, marker='o', label='Ordenado')
plt.legend()

# Adicionar título e rótulos
plt.title('Tempo de Execução do BubbleSort')
plt.xlabel('Tamanho da Entrada (n)')
plt.ylabel('Tempo de Execução (segundos)')

# Adicionar escala do eixo y
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

# Mostrar o gráfico
plt.show()
```
